# Running TrainDB

## Usage Example

```python
%%writefile tutorial.py

import os
import glob
import jaydebeapi
import sys

###################
# required setting
###################
os.environ['TRAINDB_PREFIX'] = "/content/traindb/traindb-root"
db_url = "jdbc:traindb:mysql://localhost:3306"
db_user = "root"
db_password = "root"

# create DB connection
jars = glob.glob(os.environ['TRAINDB_PREFIX'] + "/share/**/*.jar", recursive=True)
conn = jaydebeapi.connect("traindb.jdbc.Driver", db_url, [db_user, db_password], jars=jars)

# create a statement
stmt = conn.jconn.createStatement()

# CREATE MODELTYPE
stmt.execute("CREATE MODELTYPE tablegan FOR SYNOPSIS AS CLASS 'TableGAN' IN 'models/TableGAN.py'")

# SHOW MODELTYPES
rs = stmt.executeQuery("SHOW MODELTYPES")
curs = conn.cursor()
curs._rs = rs
curs._meta = curs._rs.getMetaData()
rs = curs.fetchall()
print("--- SHOW MODELTYPES")
print(rs)
curs.close()

# TRAIN MODEL
# 'epochs' is set to 1 for testing
rs = stmt.execute("TRAIN MODEL tgan MODELTYPE tablegan ON instacart.order_products(reordered, add_to_cart_order) OPTIONS ('epochs' = 1)")

# SHOW MODELS
rs = stmt.executeQuery("SHOW MODELS")
curs = conn.cursor()
curs._rs = rs
curs._meta = curs._rs.getMetaData()
rs = curs.fetchall()
print("--- SHOW MODELS")
print(rs)
curs.close()

# CREATE SYNOPSIS
stmt.execute("CREATE SYNOPSIS tgan_syn AS FILE FROM MODEL tgan LIMIT 1 PERCENT")

# SHOW MODELS
rs = stmt.executeQuery("SHOW SYNOPSES")
curs = conn.cursor()
curs._rs = rs
curs._meta = curs._rs.getMetaData()
rs = curs.fetchall()
print("--- SHOW SYNOPSES")
print(rs)
curs.close()

stmt.close()

# Query
curs = conn.cursor()
curs.execute("SELECT reordered, avg(add_to_cart_order) FROM instacart.order_products GROUP BY reordered ORDER BY reordered")
rs = curs.fetchall()
print("--- EXACT QUERY")
print(rs)

curs.execute("SELECT APPROXIMATE reordered, avg(add_to_cart_order) FROM instacart.order_products GROUP BY reordered ORDER BY reordered")
rs = curs.fetchall()
print("--- APPROXIMATE QUERY")
print(rs)

curs.close()
conn.close()
```

```python
!export PYENV_ROOT="$HOME/.pyenv" && export PATH="$PYENV_ROOT/bin:$PATH" && eval "$(pyenv init --path)" && python tutorial.py
```

## Approximate Query without Data Source

```python
%%writefile standalone.py

import os
import glob
import jaydebeapi
import sys

###################
# required setting
###################
os.environ['TRAINDB_PREFIX'] = "/content/traindb/traindb-root"
db_url = "jdbc:traindb://localhost/traindb"
db_user = ""
db_password = ""

# create DB connection
jars = glob.glob(os.environ['TRAINDB_PREFIX'] + "/share/**/*.jar", recursive=True)
conn = jaydebeapi.connect("traindb.jdbc.Driver", db_url, [db_user, db_password], jars=jars)

# create a statement
stmt = conn.jconn.createStatement()

# Approximate Query in standalone
curs = conn.cursor()
curs.execute("SELECT APPROXIMATE reordered, avg(add_to_cart_order) FROM traindb.instacart.order_products GROUP BY reordered ORDER BY reordered")
rs = curs.fetchall()
print("--- APPROXIMATE QUERY")
print(rs)

curs.close()
conn.close()
```
